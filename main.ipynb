```python
# importing Libraries
import cv2
import numpy as np
import pyautogui
```

```python
# 
while True:
    # take screenshot
    image = pyautogui.screenshot(region= (180, 200,350,300))
    # now convert pyautogui screenshot pixels to numpy array
    image = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)
    
    #COUNT PIXELS
    
    black_pixel_count = np.sum(image < 100)
    white_pixel_count = np.sum(image > 100)
    
    # print value of pixel
    print('number of black pixels : ',black_pixel_count)
    print('number of white pixels : ',white_pixel_count)
    
    if black_pixel_count > 4000 and black_pixel_count < 30000:
        pyautogui.press('up')
        
    if white_pixel_count >4000 and white_pixel_count < 30:
        pyautogui.press('up')
    
    
    
    
    # now display image
    
    cv2.imshow('image', image)
    if cv2.waitKey(25) & 0xFF == ord('q'):
        cv2.destroyAllWindows()
        break
```
